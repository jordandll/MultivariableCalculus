# Line Integrals

## Length of Curve

One use case of line integrals is to find the length of a path or curve.  

### Example

To find the length of a parabola given by $y:=f(x)=x^2$ from $(0,0)$ to $(1,1)$ we can compute the following line integral:  

$$ L = \int_C |\mathrm{d}\mathbf{r}| $$

Where,  

$$ C:\quad (x,x^2)\quad\text{for }0\le x\le 1$$

or,  

$$C:\quad \mathbf{r}:=\mathbf{r}(t) = \langle x(t),y(t)\rangle = \langle t,t^2\rangle \quad\text{for }\,0\le t\le 1 $$

Thus,  

\begin{equation}\tag{1}
    L = \int_C\sqrt{(\mathrm{d}x)^2 + (\mathrm{d}y)^2} = \int_0^1\sqrt{1+4x^2}\,\mathrm{d}x
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(0,1,num=200)
f = np.sqrt(1+4*x**2)
g = max(f)*x

fig, ((ax,ax2,ax3)) = plt.subplots(1,3,figsize=(12,5),sharey=True,width_ratios=[0.25,0.25,0.5])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
xlim = ax.set_xlim(-0.125,1.125)
ylim = ax.set_ylim(-0.125,max(f)+0.125)
ax.grid()

ax.plot(x,f,color='k',label=r'$y:=f(x)=\sqrt{1+4x^2}$')
y = np.linspace(0,1,num=20)
ax.plot([0 for i in range(len(y))],y,color='k')
ax.plot(np.linspace(0,1,num=20),[0 for i in range(20)],color='k')
ax.plot([1 for i in range(20)],np.linspace(0,max(f),num=20),color='k')
ax.legend()

ax.text(0.5,0.75,r'$R$',size=16)

ax2.set_xlabel('x-axis')
# ax2.set_ylabel('y-axis')
ax2.set_xlim(*xlim)
ax2.set_ylim(*ylim)
ax2.grid()

ax2.plot(x,f,color='k',label=r'$y:=f(x)=\sqrt{1+4x^2}$')
ax2.plot([0 for i in range(len(y))],y,color='k')
ax2.plot(np.linspace(0,1,num=20),[0 for i in range(20)],color='k')
ax2.plot([1 for i in range(20)],np.linspace(0,max(f),num=20),color='k')
ax2.plot(x,g,color='k',ls='--')

ax2.text(0.1,0.75,r'$R_1$',size=12)
ax2.text(0.6,0.75,r'$R_2$',size=12)

x = np.linspace(-1,1,num=200)
x_g = np.linspace(0,1,num=20)
f = np.sqrt(1+4*x**2)
g = max(f)*x_g
ax3.set_xlabel('x-axis')
# ax3.set_ylabel('y-axis')
ax3.set_ylim(-0.125,max(f)+0.125)
ax3.grid()

ax3.plot(x,f,color='k')
ax3.plot([0 for i in range(len(y))],y,color='k')
ax3.plot(np.linspace(0,1,num=20),[0 for i in range(20)],color='k')
ax3.plot([1 for i in range(20)],np.linspace(0,max(f),num=20),color='k')

ax3.plot(x_g,g,color='k',ls='--')
```

We can find $L$ be finding the area of the region $R$, depicted in the above figure.  The area of $R$ is of course equal to the integral found in equation (1).  We can also find $R$ and consequently $L$ with the double integral in polar coordinates:  

$$ L = \int\!\int_R\,\mathrm{d}A = \int\!\int_R r\,\mathrm{d}r\,\mathrm{d}\theta = \int\!\int_{R_1}\mathrm{d}A + \int\!\int_{R_2}\mathrm{d}A$$

where the regions $R$, $R_1$ and $R_2$ are bounded by:  

$$ R:\quad (0,y)\quad\quad (x,0)\quad\quad (1,y)\quad\quad \big(x,f(x)\big)$$

$$ R_1:\quad (0,y)\quad\quad \big(x,g(x)\big)\quad\quad \big(x,f(x)\big)$$

and  

$$ R_2:\quad (x,0)\quad\quad (1,y)\quad\quad \big(x,g(x)\big)$$

respectively.  Note that $f(x)$ is the integrand of the integral found in equation (1) and $g(x)$ gives the line that divides $R$ into the two subregions $R_1$ and $R_2$.  

$$ y:=f(x) = \sqrt{1+4x^2}\quad\quad\text{and}\quad\quad y:=g(x)=f(1)x = \sqrt{5}x$$

The area of the triangular region $R_2$ is given by:  

$$ \text{area }R_2 = \int\!\int_{R_2}\,\mathrm{d}A = 0.5bh = \frac{\sqrt{5}}{2}$$

The hyperbola given by $f(x)$ is tracked by the postition vector:  

$$ \mathbf{r} := \mathbf{r}(t) = \langle x(t),y(t)\rangle$$

$$ y^2-4x^2 = 1$$

$$ y := y(t) = \cosh t \quad\quad\text{and}\quad\quad x:=x(t) = \frac{\sinh t}{2} $$

Recall that:  

\begin{equation}\tag{3a,b}
    \cosh t = \frac{e^t+e^{-t}}{2}\quad\quad\text{and}\quad\quad \sinh t = \frac{e^t - e^{-t}}{2}
\end{equation}

$$ y^2 = \frac{e^{2t} + e^{-2t} + 2}{4} \quad\quad\text{and}\quad\quad x^2 = \frac{e^{2t} - 2 + e^{-2t}}{16}$$

Let $t_0$ and $t_1$ denote the time at which $\mathbf{r}=\langle 0,f(0)\rangle$ and $\mathbf{r}=\langle 1,f(1)\rangle$ respectively.  

$$ x(t_0)=0 \quad\implies\quad e^{t_0} = e^{-t_0} \quad\implies$$

$$ t_0=-t_0 \quad\implies\quad t_0 = 0$$

$$ y(t_0) = f(0) = 1 \quad\implies\quad e^{t_0} = 2-e^{-t_0} = 1 $$

and,  

$$ \big(x(t_1),y(t_1)\big) = \big(1,\sqrt{5}\big) \quad\implies\quad e^{t_1} - e^{-t_1} = 4\quad\text{and}\quad e^{t_1}+e^{-t_1} = 2\sqrt{5}$$

Via the above and substitution,  

$$ 4 + 2e^{-t_1}  =2\sqrt{5} \quad\therefore\quad e^{-t_1} = \sqrt{5} - 2\quad\therefore$$

$$ t_1 = -\log\big(\sqrt{5}-2\big)$$

```python
t_1 = -np.log(np.sqrt(5)-2); t_1
```

```python
t = np.linspace(-t_1,t_1,num=200)
r = (r_x,r_y) = (0.5*np.sinh(t),np.cosh(t))

fig, ax = plt.subplots(figsize=(3,3))
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ylim = ax.set_ylim(0,max(r_y)+0.125)
ax.grid()

ax.plot(*r,color='k')
```

To compute the other double integral, and thus find the area of $R_1$, let's perform a change in coordinates on the line given by $f(x)$.  

$$ y = f(x) = \sqrt{1+4x^2} \quad\therefore\quad r\sin\theta = \sqrt{1+4r^2\cos^2\theta}\quad\therefore$$

$$ r^2\sin^2\theta - 4r^2\cos^2\theta = 1 \quad\therefore\quad r^2 = \frac{1}{\sin^2\theta - 4\cos^2\theta}$$

$$ \sin^2\theta - 4\cos^2\theta = \sin^2\theta - 4 + 4\sin^2\theta  = 5\sin^2\theta - 4 $$

$$ = 1-5\cos^2\theta = -\big(\sqrt{5}\cos\theta + 1\big)\big(\sqrt{5}\cos\theta-1\big)$$

Thus,  

$$ r^2 = \frac{A}{u-1} - \frac{B}{u+1} $$

where,  

$$ A(u+1) - B(u-1) = 1 \quad\quad\text{and}\quad\quad u=\sqrt{5}\cos\theta$$

$$ Au - Bu = 0 \quad\quad\text{and}\quad\quad A+B = 1$$

Via the above,  

$$ A=B = 0.5$$

$$ r^2 := r^2(\theta)= \frac{1}{2(u-1)} - \frac{1}{2(u+1)}$$

$$ L = \int_{\theta_a}^{\theta_b}\int_0^{r(\theta)}r\,\mathrm{d}r\,\mathrm{d}\theta= 0.5\int_{\theta_a}^{\theta_b}r^2(\theta)\,\mathrm{d}\theta$$

where,  

$$ \theta_a = \arctan f(1) = \arctan\sqrt{5} \quad\quad\text{and}\quad\quad \theta_b = \pi -\theta_a$$

$$ L = \frac{1}{4}\int_{\theta_a}^{\theta_b}\frac{\mathrm{d}\theta}{\sqrt{5}\cos\theta - 1} - \frac{1}{4}\int_{\theta_a}^{\theta_b}\frac{\mathrm{d}\theta}{\sqrt{5}\cos\theta + 1} = L_1 - L_2$$

## Hyperbolic Angle

### Unit Hyperbola  

Lets say we have the (unit) hyperbola  

\begin{equation}\tag{1.1a}
    x^2-y^2 = 1
\end{equation}

\begin{equation}\tag{1.1b}
    \mathbf{r} := \mathbf{r}(t) = \langle \cosh t,\sinh t\rangle
\end{equation}

$$ x:=x(t) = \cosh t = \frac{e^t+e^{-t}}{2} \quad\quad\text{and}\quad\quad y:=y(t) = \sinh t = \frac{e^t - e^{-t}}{2}$$

$$ r^2\cos^2\theta-r^2\sin^2\theta = 1 \quad\therefore\quad r^2 = \frac{1}{\cos^2\theta-\sin^2\theta} \quad\therefore$$

$$ r^2 = \frac{1}{2\cos^2\theta-1}$$

\begin{equation}\tag{1.1c}
    r:=r(\theta) = \frac{1}{\sqrt{2\cos^2\theta - 1}}
\end{equation}

for  

$$ -\theta_\max \le \theta \le \theta_\max \quad\quad\text{and}\quad\quad \pi - \theta_\max \le \theta \le \pi + \theta_\max$$

Let $\theta_\mathrm{m}:=\theta_\max$.  

$$ 2\cos^2\theta_\max = 1 \quad\therefore\quad \cos\theta_\max = \pm\frac{1}{\sqrt{2}} \quad\therefore$$

$$ \theta_\max = \{\pm 45^\circ , \pi \pm 45^\circ\}$$

```python
def R(theta):
    return 1/np.sqrt(2*np.cos(theta)**2-1)
```

```python
x = np.concatenate((np.linspace(-4,-1,num=100),np.linspace(1,4,num=100)))
y = f = np.sqrt(x**2-1)

theta = np.linspace(-np.radians(44), np.radians(44),num=200)
r = 1/np.sqrt(2*np.cos(theta)**2-1)

alpha=np.radians(36)
g = np.tan(alpha)*x

# Create the figure and axes objects.
fig = plt.figure(figsize=(12,4.5))
ax1 = fig.add_subplot(1,4,1,projection='polar')
ax2 = fig.add_subplot(1,4,(2,3),projection=None)
ax3 = fig.add_subplot(1,4,4,projection=None)
ax1.set_xlim(-np.radians(50),np.radians(50))
ax3.set_xlim(0,2)
ax3.set_ylim(-2,2)
ax2.grid()
ax3.grid()

# Plot on axes 1 and 2.
ax1.plot(theta,r)
ax2.plot(x,y,x,-y,color='tab:blue')
ax2.plot(x,-x,x,x,color='tab:blue', ls='--')
ax2.plot(x,g,x,-g,color='tab:red',linewidth=0.8)

# Plot on axes 3.
x_2 = x[100:]
x_1 = np.linspace(0,4,num=20)
f = np.sqrt(x_2**2-1)
g = np.tan(alpha)*x_1
ax3.plot(x_1,x_1,x_1,-x_1,color='tab:blue',ls='--')
ax3.plot(x_2,f,x_2,-f,color='tab:blue')
ax3.plot(x_1,g,x_1,-g,color='tab:red',linewidth=0.8)
ax3.text(0.75,0,r'$R$',size=12)
```

Now, let's say we want to find the area of the region $R$, where $R$ is bounded by the hyperbola and:  

$$ \big(x,\pm g(x)\big)\quad\quad\text{where }\,y := g(x) = \tan\alpha x \quad\text{and}\quad \alpha = 36^\circ$$

Note that the lines described above are shaded in red in the above figure.  The aforementioned region $R$ is also labeled as such.  Now, let's find the points of intersection between these lines and the hyperbola.  Via substitution and the equations that give said curves:  

$$ x_1^2 - \tan^2\alpha x_1^2 = 1 \quad\therefore\quad x_1^2 = \frac{1}{1-\tan^2\alpha}  = \frac{\cos^2\alpha}{\cos^2\alpha - \sin^2\alpha}$$

$$ = \frac{\cos^2\alpha}{2\cos^2\alpha - 1} = \frac{\cos^2\alpha}{\cos2\alpha} \quad\therefore$$

\begin{equation}\tag{1.5a}
    x_1 = \pm\frac{\cos\alpha}{\sqrt{\cos2\alpha}}
\end{equation}

```python
x_1 = np.cos(alpha)/np.sqrt(np.cos(2*alpha)); x_1
```

Via the above equation, the equation that gives the boundary line (shaded in red), and substitution,  

\begin{equation}\tag{1.5b}
    y_1 = g(x_1) = \pm \frac{\sin\alpha}{\sqrt{\cos2\alpha}}
\end{equation}

To find the area of $R$, henceforth '$a$', let's rotate our hyperbola around the origin by $45^\circ$ (counter clockwise).  

$$ r^2 = \frac{1}{2\cos^2\big(\theta-45^\circ\big) - 1}  = \frac{1}{\cos\big(2\theta-90^\circ\big)} = \frac{1}{\sin2\theta}$$

$$ = \frac{1}{2\cos\theta\sin\theta}$$

Thus,  

$$ r\sin\theta = \frac{1}{2r\cos\theta} \quad\hat{=}\quad y = \frac{1}{2x}$$

Now, let's rotate the region $R$ and, in the process, the boundary lines thereof.  

$$ \big(x, g_1(x)\big) \quad\quad\text{where }\,y:=g_1(x) = \tan\big(45^\circ-\alpha\big)x $$

$$ \big(x, g_2(x)\big) \quad\quad\text{where }\,y:=g_2(x) = \tan\big(45^\circ+\alpha\big)x $$

```python
# Create the polar domain.
theta = np.linspace(np.radians(5),np.radians(85),num=200)
r = 1/np.sqrt((2*np.cos(theta-0.25*np.pi)**2-1))

# Create the rectilinear domain.
x = np.linspace(1/(2.5*2),2.5,num=200)
x_maj = np.linspace(0,max(x),num=20)

y = 1/(2*x)
g_1 = np.tan(-alpha+0.25*np.pi)*x_maj
g_2 = np.tan(alpha+0.25*np.pi)*x_maj

# Create the figure and axes objects.
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1,projection='polar')
ax2 = fig.add_subplot(1,2,2,projection=None)
xlim = ax1.set_xlim(0,0.5*np.pi)
ax2.set_ylim(-0.1,max(y)+0.1)
ax2.grid()

# Plot on the 2 axes.
ax1.plot(theta,r)
ax2.plot(x,y)
ax2.plot(np.linspace(0,max(x),num=20),[0 for i in range(20)],color='tab:blue',ls='--')
ax2.plot([0 for i in range(20)],np.linspace(0,max(y),num=20),color='tab:blue',ls='--')

ax2.plot(x_maj,x_maj,color='grey',linewidth=0.8)
ax2.plot(x_maj,g_1,x_maj,g_2,color='tab:red',linewidth=0.8)
```

The new points of intersection between the red boundary lines and the hyperbola found in the above figure are given by:  

$$ \tan\big(45^\circ-\alpha\big) = \frac{1}{2x_\mathrm{a}^2}\quad\therefore\quad x_\mathrm{a}^2 = 0.5\cot\big(45^\circ-\alpha\big) \quad\therefore $$

$$x_\mathrm{a} = \pm\sqrt{0.5\cot\big(45^\circ-\alpha\big)}$$

$$ y_\mathrm{a} = g_1(x_\mathrm{a}) = \pm\sqrt{0.5\tan\big(45^\circ-\alpha\big)}$$

As for the points of intersection between the line given by $g_2(x)$ and the hyperbola:  

$$ x_\mathrm{b}^2 = 0.5\cos\big(45^\circ+\alpha\big) \quad\therefore\quad x_\mathrm{b} = \pm\sqrt{0.5\cot\big(45^\circ+\alpha\big)}$$

$$ y_\mathrm{b} = g_2(x_\mathrm{b}) = \sqrt{0.5\tan\big(45^\circ+\alpha\big)}$$

The area of $R$ is thus:  

$$ a = \text{area }R = \int_0^{x_\mathrm{b}}g_2(x)\,\mathrm{d}x - \int_0^{x_\mathrm{b}}g_1(x)\,\mathrm{d}x + \frac{1}{2}\int_{x_\mathrm{b}}^{x_\mathrm{a}}\frac{\mathrm{d}x}{x} - \int_{x_\mathrm{b}}^{x_\mathrm{a}}g_1(x)\,\mathrm{d}x$$

$$ = \int_0^{x_\mathrm{b}}g_2(x)\,\mathrm{d}x - \int_0^{x_\mathrm{a}}g_1(x)\,\mathrm{d}x + \frac{1}{2}\int_{x_\mathrm{b}}^{x_\mathrm{a}}\frac{\mathrm{d}x}{x}$$

\begin{equation}\tag{1.2}
    a=0.5\tan\big(45^\circ+\alpha\big)x_\mathrm{b}^2 - 0.5\tan\big(45^\circ-\alpha\big)x_\mathrm{a}^2 + 0.5\big(\log(x_\mathrm{a}) - \log(x_\mathrm{b})\big)
\end{equation}

$$ = 0.5\log\frac{x_\mathrm{a}}{x_\mathrm{b}} = \frac{1}{4}\,\log\Big(\cot\big(45^\circ-\alpha\big)\tan\big(45^\circ+\alpha\big)\Big)$$

$$ = \frac{1}{4}\,\log\Big(\cot\big(45^\circ-\alpha\big)\Big) + \frac{1}{4}\log\Big(\tan\big(45^\circ+\alpha\big)\Big)$$

$$ = \frac{1}{4}\Big(\log\cos\big(45^\circ-\alpha\big) - \log\sin\big(45^\circ-\alpha\big) + \log\sin\big(45^\circ+\alpha\big) - \log\cos\big(45^\circ+\alpha\big)\Big)$$

It's seems impossible to compute the left hand integral of the above expression, so let's try a change in variables.

And, the numerator of said fraction "simplifies" to:

The denominator of the above fraction expands and "simplifies" to:

The area element is given by:  

$$ \mathrm{d}A = \left|\frac{\partial(x,y)}{\partial(u,v)}\right|\,\mathrm{d}u\,\mathrm{d}v$$

The partial derivatives that make up the Jacobian found in the above equation are, via the equation that define $u(x,y)$ and $v(x,y)$, given by:  

$$ v = (u-x)^2-x^2 = u^2 - 2ux + x^2 - x^2 = u^2 - 2ux \quad\therefore$$

$$ \frac{\partial(x,y)}{\partial(u,v)} = \left|\begin{matrix}x_u & x_v \\ y_u & -x_v\end{matrix}\right| = x_v\left|\begin{matrix}x_u & 1 \\ y_u & -1\end{matrix}\right|$$

$$ \left|\frac{\partial(x,y)}{\partial(u,v)}\right| = -x_v\big(x_u+y_u\big) = -x_v = \frac{1}{2u}$$

$$ a = \int\!\int_R\,\mathrm{d}A = \int_0^1\int_{h_1(v)}^{h_2(v)}\frac{\mathrm{d}u\,\mathrm{d}v}{2u}$$
