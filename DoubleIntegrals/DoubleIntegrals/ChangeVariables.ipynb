### Problem 1

Compute $\displaystyle \int\!\int_R\left(\frac{x+y}{2-x+y}\right)^4\,\mathrm{d}x\,\mathrm{d}y$, where $R$ is the square with vertices at $(1,0)$, $(0,1)$, $(-1,0)$, and $(0,-1)$.

Let $u$ and $v$ be defined as:  

\begin{equation}\tag{1.1a,b}
    u:= x+y \quad\quad\text{and}\quad\quad v:=y-x
\end{equation}

Via the above two equations,  

$$ v = y - u + + y = 2y-u$$

\begin{equation}\tag{1.1c}
    y = \frac{u+v}{2}
\end{equation}

And, via the above equation and equation (1.1a) and substitution,  

\begin{equation}\tag{1.1d}
    x = u - \frac{u+v}{2} = \frac{u-v}{2}
\end{equation}

The region $R$ is bounded by:  

\begin{matrix}
y &= 1-x \quad\quad &y = x-1 \\
y &= 1+x \quad\quad &y = -1-x
\end{matrix}

Or,  

$$ x+y = u =\pm 1 \quad\quad\text{and}\quad\quad y-x = v = \pm 1$$

For $-1\le x,y\le1$.

```python
x = np.linspace(-1,1,num=50)
X, Y = np.meshgrid(x,x)
f = ((X+Y)/(2-X+Y))**4
u = X+Y
v = Y-X

fig, ax = plt.subplots(figsize=(8,5))
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

ax.plot(x[:25], 1+x[:25], x[:25], -x[:25] - 1, color='k', ls='--')
ax.plot(x[25:], 1-x[25:], x[25:], x[25:] - 1, color='k', ls='--')
cs = ax.contour(X,Y,f,levels=list(np.linspace(0,0.15,num=12))+[0.2,0.3],cmap='terrain')
cb = fig.colorbar(cs)
cs_u = ax.contour(X,Y,u,colors='lightgrey',linewidths=1.0,levels=list(np.linspace(-np.sqrt(2),np.sqrt(2),num=8)))
ax.contour(X,Y,v,colors='lightgrey',linewidths=1.0,levels=list(np.linspace(-np.sqrt(2),np.sqrt(2),num=8)))
cb_u = fig.colorbar(cs_u)
```

The area element is given by:  

$$ \mathrm{d}A = \left|\frac{\partial(x,y)}{\partial(u,v)}\right|\,\mathrm{d}u\,\mathrm{d}v$$

where,  

$$ \frac{\partial(x,y)}{\partial(u,v)} = \left|\begin{matrix} x_u & x_v \\  y_u & y_v\end{matrix}\right|$$

\begin{equation}\tag{1.3}
    = x_uy_v - x_vy_u = 0.5
\end{equation}

$$ x_u = y_u = y_v = 0.5 \quad\quad\text{and}\quad\quad x_v = -0.5$$

Note that,  

$$ \frac{\partial(u,v)}{\partial(x,y)} = \left|\begin{matrix} u_x & u_y \\ v_x & v_y\end{matrix}\right| = 2$$

$$ u_x = u_y = v_y = 1 \quad\quad\text{and}\quad\quad v_x = -1$$

Thus, under the new $u,v$ coordinate system, the double integral is:  

$$ 0.5\int_{-1}^1\int_{-1}^1\left(\frac{u}{2+v}\right)^4\,\mathrm{d}u\,\mathrm{d}v$$

The inner integral is:  

$$ \frac{2}{5\big(2+v\big)^4}$$

Thus,  

$$ \frac{1}{5}\int_{-1}^1\frac{1}{\big(2+v\big)^4}\,\mathrm{d}v$$

$$ = -\left.\frac{1}{15}\cdot\frac{1}{\big(2+v\big)^3}\right\vert_{\,-1}^{\,1}$$

$$ = \frac{1}{15}\left(1-\frac{1}{27}\right) = \frac{1}{15}\cdot\frac{26}{27}$$
