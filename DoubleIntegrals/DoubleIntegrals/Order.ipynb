# Order of Integration

### Example 1  

Evaluate  

$$ I = \int_0^{0.5\pi}\int_x^{0.5\pi} \frac{\sin y}{y}\,\mathrm{d}y\,\mathrm{d}x$$

by changing the order of integration.

Let $f(x,y)$ be defined as:  

$$ f(x,y) = \frac{\sin y}{y}$$

$$ I = \int\!\int_R f(x,y)\,\mathrm{d}A$$

where the region $R$ is bounded by:  

$$ 0 \le x,y\le 0.5\pi \quad\quad\text{and}\quad\quad y = x$$

Another way to describe $R$ is that it is the triangle whose vertices are equal to the intersection of the lines given by:  

$$ (0,y) \quad\quad (x,0.5\pi) \quad\quad (x,y)$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = np.pi
sin = np.sin
x = y = np.linspace(0,0.5*pi,num=50)
g = x
X, Y = np.meshgrid(x,y)
f = sin(Y)/Y

fig, ax = plt.subplots()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

ax.plot(x,g,ls='--',color='k')
ax.plot([0 for i in range(len(y))],y,ls='--',color='k')
ax.plot(x,[0.5*pi for i in range(len(x))],ls='--',color='k')
cs = ax.contourf(X,Y,f,levels=12)
cb = fig.colorbar(cs)
```

#### Pre Change in Order
First, let's compute the double integral that is equal to $I$ *without* changing the order of integration.  The first iteration of the computation is:  

$$ \int_x^{0.5\pi} f(x,y)\,\mathrm{d}y = \text{???}$$

As it turns out, finding the primitive integral of $f(x,y)$ with respect to $y$ appears to be impossible!

#### Post Change in Order

To make computing the double integral possible in the first place, let's change the order of integration.  The inner integral is now with respect to $x$ and the limit of said integral is from $0$ to $y$.  

$$ \int_0^y f(x,y)\,\mathrm{d}x = \big(f(x,y)\cdot x\big)\Big\vert_{\,0}^{\,y} = \sin y$$

The outer integral is with respect to $y$ and the limit thereof is from $0$ to $0.5\pi$.  

$$ I = \int_0^{0.5\pi} \sin y \,\mathrm{d}y = -\cos y \,\Big\vert_{\,0}^{\,0.5\pi} = 1$$

### Problem 1

Calculate  $\displaystyle \int_0^2\int_x^2 e^{-y^2}\,\mathrm{d}y\,\mathrm{d}x$.

Let $f(x,y):=e^{-y^2}$.  As is, the inner integral may be impossible to compute without some numerical method, since the primitive integral of $f(x,y)$ with respect to $y$ seems impossible to find.  Pre-change in order of integration, the inner integral is with respect to $y$ and the limit thereof is from $x$ to $2$, while the limit of the outer integral is from $0$ to $2$.

Changing the order of integration we get,  

$$ I:=\int_0^2\int_0^y e^{-y^2}\,\mathrm{d}x\,\mathrm{d}y$$

The inner integral is given by:  

$$ \int_0^y e^{-y^2} \,\mathrm{d}x = xe^{-y^2}\,\Big\vert_{\,0}^{\,y} = ye^{-y^2}$$

The outer integral is thus:  

$$ I = \int_0^2 ye^{-y^2}\,\mathrm{d}y = -0.5e^{-y^2}\,\Big\vert_{\,0}^{\,2}$$

$$ = 0.5 - 0.5e^{-4} = 0.5\big(1-e^{-4}\big)$$
