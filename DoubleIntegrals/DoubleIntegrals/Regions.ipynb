```python
x = np.linspace(0,1,num=50)
y = np.linspace(0,2,num=50)

f = x+1
g = x**2

X,Y = np.meshgrid(x,y)

d = X*Y

fig, ax = plt.subplots()
ax.set_title('Density',size=13)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

ax.plot(x,f,ls = '--', color='k', label = r'$f(x)=x+1$')
ax.plot(x,g,ls = '--', color='k', label = r'$g(x)=x^2$')
ax.plot([1 for i in range(len(y))],y,ls='--',color='k')
cs = ax.contour(X,Y,d,levels=12)
cb = fig.colorbar(cs)
```

The mass is given by:  

$$ m = \int\!\int_R xy\,\mathrm{d}A = \int_0^1\int_{x^2}^{x+1} xy\,\mathrm{d}y\mathrm{d}x$$

The first iteration is:  

$$ \left.\int xy \,\mathrm{d}y\,\right|_{x^2}^{x+1} = \left. 0.5xy^2\,\right|_{x^2}^{x+1} = 0.5x\big(x^2+2x+1 - x^4\big)$$

Thus,  

$$ m = \int_0^1 0.5x\big(x^2+2x+1-x^4\big)\,\mathrm{d}x$$

$$ = 0.5\left.\left(0.25x^4 + \frac{2}{3}\,x^3 + 0.5x^2 - \frac{x^6}{6}\right)\right|_0^1$$

$$ = 0.50\left(\frac{3}{4} + \frac{2}{3} - \frac{1}{6}\right)$$

$$ = 0.5\big(0.75 + 0.5\big) = 0.5\cdot1.25 = \frac{1}{2}\,\frac{5}{4} = \frac{5}{8}$$

#### Problem 2  

First, let's find the equation of the plane $\mathcal{P}$ that is diagonal to the coordinate planes and one of the faces of the tetrahedron rests.  Let $\mathbf{a}$ and $\mathbf{b}$ be defined as:  

$$ \mathbf{a} := \langle 0,1,0\rangle - \langle 1,0,0\rangle = \langle -1,1,0\rangle \quad\quad \text{and}\quad\quad \mathbf{b} := \langle 0,0,1\rangle - \langle 1,0,0\rangle = \langle -1,0,1\rangle$$

Whereas the cross product of the vectors defined above is normal to $\mathcal{P}$ therefore the equation that gives $\mathcal{P}$, in point normal form, is:  

$$ c_xx + c_yy + c_z(z-1) = 0$$

where,  

$$ \mathbf{c}:= \mathbf{a}\times\mathbf{b} = \left|\begin{matrix}\mathbf{i} & \mathbf{j} & \mathbf{k} \\ -1 & 1 & 0 \\ -1 & 0 & 1 \end{matrix}\right|$$

$$ = \mathbf{i}  + \mathbf{j} + \mathbf{k}$$

Thus, $\mathcal{P}$ is given by:    

$$ x + y + z = 1$$

$$ z := f(x,y) = 1-x-y$$

The volume of the tetrahedron is given by:  

$$ V = \int\!\int_R f(x,y)\,\mathrm{d}A$$

Here, the region $R$ is bounded by:  

$$ 0\le x,y \le 1 \quad\quad\text{and}\quad\quad x+y=1$$

Thus,  

$$ V = \int_0^1\int_0^{1-x} f(x,y)\,\mathrm{d}y\,\mathrm{d}x$$

The first iteration is:  

$$ \int_0^{1-x}f(x,y)\,\mathrm{d}y = \big(y(1-x) - 0.5y^2\big)\Big\vert_0^{1-x}$$

$$ = 0.5(1-x)^2 = 0.5\big(1 - 2x + x^2\big)$$

$$ V = \int_0^1 0.5\big(1-2x+x^2\big)\,\mathrm{d}x$$

$$ = 0.5\left.\left(x - x^2 + \frac{x^3}{3}\right)\right\vert_0^1$$

$$ = \frac{1}{6} $$
