### Problem 2  

Consider the vector field $\displaystyle \mathbf{F} = -c\frac{x\,\mathbf{i}+y\,\mathbf{j}}{x^2+y^2}$.  (This is the gravitational field of an infinitely long uniform wire along the $z$-axis.)  Take $c=1$ and find, by direct calculation (i.e. using a parametrization of $C$), the work done by the gravitational field in moving a unit mass along each of the following paths $C$ in the $xy$-plane.

Assuming $c=1$,  

$$ \mathbf{F} = -\frac{\langle x,y\rangle}{r^2} \quad\quad\text{where }\, r = \sqrt{x^2+y^2}$$

**a)**  $C$ is the half-line $y=1$ for $x\ge 0$.

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = -\int_C \frac{1}{r^2}\big(x\,\mathrm{d}x + y\,\mathrm{d}y\big)$$

$$ y = 1 = \mathit{const}\quad\therefore\quad \mathrm{d}y = 0$$

Thus,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = -\int_0^\infty\frac{x}{x^2+1}\,\mathrm{d}x$$

$$ = \left.-0.5\log\big(x^2+1\big)\right]_{\,0}^{\,\infty} = \left.\log\frac{1}{\sqrt{x^2+1}}\right]_{\,0}^{\,\infty} = -\infty$$

```python
np.log(np.inf)
```

**b)**  $C$ is the circle of radius $a$ with center at origin, traced counterclockwise.

$$ \mathbf{r} := \mathbf{r}(t) = a\langle \cos t, \sin t\rangle = \langle x,y\rangle\quad\quad\text{and}\quad\quad a = r$$

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = -\frac{1}{a}\int_C \cos t\,\mathrm{d}x + \sin t\,\mathrm{d}y$$

$$ x'(t) = -a\sin t \quad\therefore\quad \mathrm{d}x = -a\sin t\,\mathrm{d}t$$

$$ y'(t) = a\cos t \quad\therefore\quad \mathrm{d}y = a\cos t\,\mathrm{d}t$$

Finally,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = -\int_0^{2\pi}\cos t\sin t\,\mathrm{d}t - \cos t \sin t\,\mathrm{d}t = 0$$

Note that, we could have inferred the above from the Fundamental Theorem of line integrals, the fact that $\mathbf{F}$ is conservative, and the fact that $C$ is a closed curve, but in doing so, we would not be solving the problem via direct calculation.

**c)** $C$ is the line segment from $(0,1)$ to $(1,0)$.

The path rests on $y=1-x$ therefore $\mathrm{d}y = -\mathrm{d}x$.  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = -\int_C\frac{1}{r^2}\big(x\,\mathrm{d}x + y\,\mathrm{d}y\big)$$

$$ = -\int_0^1\frac{2x - 1}{2x^2+1 - 2x}\,\mathrm{d}x$$

$$ = \left.-0.5\log\big(2x^2+1-2x\big)\right]_{\,0}^{\,1} = 0$$

To verify the above, we note that the potential of $\mathbf{F}$, denoted as '$f(x,y)$', is given by:  

$$ \mathbf{F} = \nabla f \quad\iff\quad f(x,y) = -\log\sqrt{x^2+y^2} = -\log r$$

Via the Fundamental Theorem for line integerals,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = f(1,0) - f(0,1) = 0$$

### Problem 4  

Let $\mathbf{F} = \nabla\big(x^2y + 2xy^2\big)$ and $C$ be the quarter ellipse $9x^2+4y^2=1$ running from the positive $x$-axis to the positive $y$-axis.  

**a)**  Write the integral $\displaystyle \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r}$ in the form $\displaystyle \int_CM\,\mathrm{d}x + N\,\mathrm{d}y$.  

Let $\mathbf{F} := \nabla f$.  

$$ M = f_x = 2xy + 2y^2 \quad\quad\text{and}\quad\quad N = x^2 + 4xy$$

Thus,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = \int_C\big(2xy+2y^2\big)\,\mathrm{d}x + \big(x^2+4xy\big)\,\mathrm{d}y$$

**b)**  By parametrizing the curve $C$, write the integral $\displaystyle\int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r}$ as an explicit definite integral in $t$.  (*Do not evaluate*).  

$$ \mathbf{r} := \mathbf{r}(t) = \langle x, y\rangle = \left\langle \frac{\cos t}{3}, \frac{\sin t}{2}\right\rangle  $$

$$  x:= x(t) = \frac{\cos t}{3} \quad\quad\text{and}\quad\quad y := y(t) = \frac{\sin t}{2}$$

$$ x'(t) = -\frac{\sin t}{3}\quad\therefore\quad \mathrm{d}x = - \frac{\sin t}{3}\,\mathrm{d}t$$

$$ \mathrm{d}y = \frac{\cos t}{2}\,\mathrm{d}t$$

The $x$ and $y$ compliments of the field, in terms of $t$ are thus,  

$$ M = \frac{\sin2t}{6} + \frac{\sin^2 t}{2} \quad\quad\text{and}\quad\quad N = \frac{\cos^2 t}{9} + \frac{\sin 2t}{3}$$

The integrand of our line integral, in terms of $t$, is:  

$$ M\,\mathrm{d}x + N\,\mathrm{d}y = \left(\frac{\cos^3 t}{18}+ \frac{\sin t\cos^2 t}{6}-\frac{\sin^2 t\cos t}{3} - \frac{\sin^3 t}{6}\right)\,\mathrm{d}t$$

The range of the integral in $t$ is from $0$ radians to $0.5\pi$ radians.

**c)**  Via the Fundamental Theorem for line integrals,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = f(x_1,y_1) - f(x_0,y_0)$$

where  

$$ y_0 = 0 \quad\quad x_0 = \frac{1}{3}\quad\quad x_1 = 0 \quad\quad y_1 = \frac{1}{2}$$

Finally,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = 0$$

**d)**  Via path independence and the fact that $\mathbf{F}$ is conservative,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = \int_{C_1}\mathbf{F}\cdot\mathrm{d}\mathbf{r_1} + \int_{C_2}\mathbf{F}\cdot\mathrm{d}\mathbf{r_2}$$

Here $C_1$ is the directed line segment from $(x_0,0)$ to $(0,0)$, which is tracked by the position vector $\mathbf{r_1} = \langle x,0\rangle$.  

$$ \int_{C_1}\mathbf{F}\cdot\mathrm{d}\mathbf{r_1} = \int_{x_0}^0M\,\mathrm{d}x = 0$$

And $C_2$ is the directed line segment from $(0,0)$ to $(0,y_1)$, which is tracked by $\mathbf{r_2} = \langle 0,y\rangle$.  

$$ \int_{C_2}\mathbf{F}\cdot\mathrm{d}\mathbf{r_2} = \int_0^{y_1} N\,\mathrm{d}y = 0$$

Finally, via the above three equations,  

$$ \int_C\mathbf{F}\cdot\mathrm{d}\mathbf{r} = 0$$

```python
# Define some parameters.
x_0 = 1/3; y_1 = 0.5

# Create the domains.
t = np.linspace(0,2*np.pi,num=200)
x = np.linspace(-x_0,x_0,num=200)

# Define the curve C.
f = 0.5*np.sqrt(1-9*x**2)
r = (r_x,r_y) = (np.cos(t)/3,np.sin(t)/2)

# Create the figure and axes objects.
fig, ax = plt.subplots(figsize=(4.5,7))
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

# Plot the curve C.
ax.plot(x,f,x,-f,color='k')
ax.plot(*r,color='tab:blue',ls='--')
```

### Problem 5  

**a)**  Whereas the field is continuously differentiable throughout the $xy$-plane and    

$$ M_y = x \quad\quad\text{and}\quad\quad N_x = 3x^2$$  

therefore $\mathbf{F}$ is non-conservative throughout the $xy$-plane.

**b)**  Trying to apply method 1 from V2.2-3 we get:  

$$ f(x_1,y_1) = \int_{(0,0)}^{(x_1,y_1)}\mathbf{F}\cdot\mathrm{d}\mathbf{r} = \int_{C_1}\mathbf{F}\cdot\mathrm{d}\mathbf{r_1} + \int_{C_2}\mathbf{F}\cdot\mathrm{d}\mathbf{r_2}$$

where $C_1$ is the directed line segment from $(0,0)$ to $(x_1,0)$, which is tracked by $\mathbf{r_1}=\langle x,0\rangle$.  

$$ \int_{C_1}\mathbf{F}\cdot\mathrm{d}\mathbf{r_1} = \int_0^{x_1}xy\,\mathrm{d}x = 0$$

and $C_2$ is the directed line segment from $(x_1,0)$ to $(x_1,y_1)$, which is tracked by $\mathbf{r_2} = \langle x_1,y\rangle$.  The fact that $x = x_1 = \mathit{const}$ implies that $\mathrm{d}x = 0$.  

$$ \int_{C_2}\mathbf{F}\cdot\mathrm{d}\mathbf{r_2} = \int_0^{y_1} x_1^3\,\mathrm{d}y = y_1x_1^3$$

Via the above three equations, substitution, dropping the subscripts, and adding $c$, an arbitrary constant of integration,  

$$ f(x,y) = yx^3 + c$$

However,  

$$ \nabla f = \langle 3x^2y, x^3\rangle \ne \mathbf{F}$$

**c)**  Trying to apply method 2 from V2.2-3 we get:  

$$ f_x = xy \quad\implies\quad f = \int xy\,\mathrm{d}x = 0.5x^2y + g(y)$$

$$ f_y = x^3 \quad\quad\text{and}\quad\quad f_y = 0.5x^2 + g'(y)$$

Via the above two equations and equivalence principle,  

$$ g'(y) = x^3 - 0.5x^2 \quad\implies\quad g(y) = \int\big(x^3-0.5x^2\big)\,\mathrm{d}y$$

$$ g(y) = \big(x^3-0.5x^2\big)y + c$$

The (false) potential we get using method 2 is:  

$$ f(x,y) = x^3y + c$$

Which is the same one we got using method 1, which is verifiably not the potential of our field.
