# Plane Vector Fields

## Force and Velocity Fields

### Force Fields  

From physics, we have the two-dimensional electrostatic force fields arising from a distri
bution of static (i.e., not moving) charges in the plane. At each point (x0, y0) of the plane,
we put a vector representing the force which would act on a unit positive charge placed at
that point.  

In the same way, we get vector fields arising from a distribution of masses in the xy-plane,
representing the gravitational force acting at each point on a unit mass. There are also the
electromagnetic fields arising from moving electric charges and/or a distribution of magnets,
representing the magnetic force at each point.
Any of these we shall simply refer to as a **force field**.

#### Example 2

Express in $\mathbf{i} − \mathbf{j}$ form the electrostatic force field $\mathbf{F}$ in the $xy$-plane arising from a unit positive charge placed at the origin, given that the force vector at $(x, y)$ is directed radially away from the origin and that it has magnitude $c/r^2$ , $c$ constant.

Let $\mathbf{r}$ be defined as:  

$$ \mathbf{r} := \mathbf{r}(x,y) := \langle x,y\rangle \quad\implies\quad |\mathbf{r}| = r = \sqrt{x^2+y^2}$$

The direction of the abover vector is the same as our force field.  

$$ \text{dir }\mathbf{r} = \text{dir }\mathbf{F}$$

Thus,  

$$ \mathbf{F} = A\mathbf{r} $$

where $A$ is some constant scalar value that we can find with the following equation:  

$$ |\mathbf{F}| = A|\mathbf{r}| = Ar = \frac{c}{r^2}\quad\therefore\quad A = \frac{c}{r^3}$$

Finally,  

$$ \mathbf{F} = \frac{c}{r^3}\,\langle x,y\rangle \quad\therefore\quad |\mathbf{F}| = \frac{c}{r^2}$$

### Flow and Velocity Fields

There is the **velocity field** $\mathbf{v}(x, y)$ where the vector $\mathbf{v}(x, y)$ at the point $(x, y)$ represents the velocity vector of the flow at that point — that is, its direction gives the direction of flow, and its magnitude gives the speed of the flow.  

Then there is the **flow field**, defined by  

\begin{equation}\tag{6}
	\mathbf{F} = \delta(x,y)\mathbf{v}(x,y)
\end{equation}

where $\delta(x, y)$ gives the density of the fluid at the point $(x, y)$, in terms of mass per unit area. Assuming it is not $0$ at a point $(x, y)$, we can interpret $\mathbf{F}(x, y)$ as follows:  

* $\text{dir }|\mathbf{F}|$ is the direction of the fluid flow at $(x,y)$;  
* $|\mathbf{F}|$ is the rate (per unit length per second) of mass transport across a line that is perpendicular to the direction of the flow, denoted as '$\text{dir }|\mathbf{F}|$', at $(x,y)$. 

![flow-field-ex](FlowFieldEx.svg)

Namely, we see that first by (6) and then by the picture,  

$$ |\mathbf{F}|\Delta l \Delta t = \delta|\mathbf{v}|\Delta t\Delta l = \text{mass in }\Delta A$$  

from which (7) follows by dividing by $\Delta l \Delta t$ and letting $\Delta l$ and $\Delta t \to 0$.  

If the density is a constant $\delta_0$, as it would be for an incompressible fluid at a uniform temperature, then the flow field and velocity field are essentially the same, by (6) — the vectors of one are just a constant scalar multiple of the vectors of the other.

#### Example 3

Describe and interpret $\displaystyle \mathbf{F} = \frac{\langle x, y\rangle}{x^2+y^2}$ as a flow field and a force field.

As a **flow field** it is directed radially outward as in Example 2, but the magnited is greater.  

\begin{equation}\tag{3.1}
    |\mathbf{F}| := |\mathbf{F}|(x,y)= \frac{1}{\sqrt{x^2+y^2}} = \frac{1}{r}
\end{equation}

Suppose we have a circle with a radius of $a$ centered around origin.  At every point $P$ on this circle, the rate of mass transport is given by:  

$$ |\mathbf{F}|(P) = \frac{1}{a}$$

And, the rate of mass transport across the circle, henceforth $\mathcal{C}$, is given by:  

$$ |\mathbf{F}|(\mathcal{C}) = \frac{1}{a}\cdot 2\pi \cdot a = 2\pi$$

To elaborate on the above equation: the circle $\mathcal{C}$ is given by:  

$$ x^2+y^2 = a^2 \quad\therefore\quad y := f(x) = \sqrt{a^2-x^2} \quad\text{and}\quad y = -f(x)$$

$$ $$

Via the above and equation (3.1) and substitution, we obtain and equation and function that gives the magnitude of the flow field constrained to our circle.

$$ |\mathbf{F}|(P) = |\mathbf{F}|\big(x,f(x)\big) = \frac{1}{a}$$

Now let $s$, be defined as:  

$$ s:= s(x,y) := \theta r \quad\quad\text{where } \theta := \theta (x,y) = \arccos \frac{x}{r}$$

Thus,  

$$ s(P) = \theta(P)r = a\arccos \frac{x}{a}$$

FInally,  

$$ |\mathbf{F}|(\mathcal{C}) = \int_0^{2\pi a} \frac{\mathrm{d}s(P)}{a} = 2\pi$$

#### Problem 1

Sketch the following vector fields. Pay attention to their names because we will be
encountering these fields frequently.  

**a)**  Force, constant gravitational field $\mathbf{F}(x,y) = -g\mathbf{j}$.  

```python
x = np.linspace(-1,1,num=10)
x, y = np.meshgrid(x,x)
g = 9

F_x = [[0 for j in range(len(x))] for i in range(len(x))]
F_y = [[-g for j in range(len(x))] for i in range(len(x))]
F = (F_x, F_y)

fig, ax = plt.subplots(figsize=(3,3))
ax.set_title('Constant Gravitational Field')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.quiver(x,y,*F)
```

**b)**  Velocity $\displaystyle \mathbf{V}(x,y) = \frac{x\mathbf{i} + y\mathbf{j}}{x^2+y^2} = \frac{\langle x,y\rangle}{r^2}$.  (This is a shrinking radial field -- like water pouring from a source at $(0,0)$.)  

```python
V_x = x/(x**2+y**2)
V_y = y/(x**2+y**2)
V = (V_x,V_y)

fig, ax = plt.subplots(figsize=(5,5))
ax.set_title(r'Pouring Water at $(0,0)$')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.quiver(x,y,*V)
```

**c)**  Unit tangential field $\mathbf{F} = \langle -y,x\rangle/r$.  

```python
r = np.sqrt(x**2+y**2)
F = (-y/r,x/r)

fig, ax = plt.subplots(figsize=(3,3))
ax.set_title(r'$\mathbf{F} = \langle -y,x\rangle/r$')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.quiver(x,y,*F)
```

**d)**  Gradient $\mathbf{F} = \nabla f$, where $\displaystyle f(x,y) = \frac{xy}{3}$ and $\displaystyle \nabla f = \frac{\langle y,x\rangle}{3}$.  

```python
F = (y/3,x/3)
f = x*y/3

fig, ax = plt.subplots(figsize=(5,5))
ax.set_title(r'Gradient $\mathbf{F} = \nabla f$')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

cmap_f = plt.colormaps['coolwarm']
cs = ax.contour(x,y,f,cmap=cmap_f.reversed(),vmin = -1/3, vmax = 1/3, levels=12)
cb = fig.colorbar(cs)
ax.quiver(x,y,*F)
```
