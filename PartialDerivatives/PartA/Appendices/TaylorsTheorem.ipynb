# Taylor Approximation
## Univariate
### Example 1
Let $f(x)$ be defined as:  

$$f(x):=y :=\frac{1}{1+x^2}$$

$$ f'(x) = \frac{\mathrm{d}f}{\mathrm{d}x} = -\frac{2x}{\big(1+x^2\big)^2} \quad\quad f''(x) = -\frac{2}{\big(1+x^2\big)^2} + \frac{8x^2}{\big(1+x^2\big)^3}$$

$$ f''(x) = \frac{8x^2 - 2\big(1+x^2\big)}{\big(1+x^2\big)^3} = \frac{6x^2-2}{\big(1+x^2\big)^3}$$

```python
def f_x(x):
    return -2*x/(1+x**2)**2
def f_xx(x):
    return (6*x**2-2)/(1+x**2)**3
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-2,2,num=200)
y = f = 1/(1+t**2)
P_1 = 1-0.5*t
P_2 = 0.5*t**2 - 1.5*t + 1.5

fig, ax = plt.subplots()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_ylim(-0.05,1.1)

ax.plot(t,y,label=r'$f(x)$')
ax.plot(t,P_1,label=r'$P_1(x)$')
ax.plot(t,P_2,label=r'$P_2(x)$')
ax.legend()
```

#### Linear Approximation

$$ P_1(x) = f(a) + f'(a)(x-a)$$

Assuming $a=1$,  

```python
x = sp.symbols('x')
P_1 = 0.5 + f_x(1)*(x-1); P_1
```

#### Quadratic Approximation

$$ P_2(x) = f(a) + f'(a)(x-a) + 0.5f''(a)(x-a)^2 = P_1(x) + 0.5f''(a)(x-a)^2$$

$$ = 0.5\big(x^2-3x+3\big)$$

```python
P_2 = P_1 + f_xx(1)*(x-1)**2; P_2
```

```python
P_2.expand()
```

## Bivariate

### Example 1

Let $f(\mathbf{x})$ be defined as:  

$$ f(\mathbf{x}) := f(x,y) := z = x_1^2 + x_2^2 = x^2+y^2$$

The above funtion of course gives a conic parabola.  

$$ z_x = f_x = \frac{\partial z}{\partial x} = 2x \quad\quad \text{and}\quad\quad z_y = 2y$$

$$ z_{xx} = 2 \quad\quad z_{xy} = z_{yx} = 0 \quad\quad z_{yy} = 2$$

#### Linear Approximation

In general, the linear (Taylor) approximation of $f(\mathbf{x})$ at $\mathbf{a}=(a_1,a_2)=(a_x,a_y)$ is given by:  

$$ P_1(\mathbf{x}\,|\,\mathbf{a}) = f(\mathbf{a}) + f_x(\mathbf{a})v_x + f_y(\mathbf{a})v_y \quad\quad\text{where }\mathbf{v} = \mathbf{x}-\mathbf{a}$$

Note the indexing or subscripting notation used in the above equation says that $v_1=v_x=x_1-a_1=x-a_x$.

Now, assuming $\mathbf{a} = (1,1)$,  

$$ P_1(\mathbf{x}):=P_1\big(\mathbf{x}\,|\,(1,1)\big) = 2 + 2(x-1) + 2(y-1)$$

$$ = 2(x+y-1)$$

#### Quadratic Approximation

In general, the quadratic (Taylor) approximation of $f(\mathbf{x})$ at $\mathbf{a}$ is given by:  

$$ P_2(\mathbf{x}\,|\,\mathbf{a}) = P_1(\mathbf{x}\,|\,\mathbf{a}) + 0.5f_{xx}(\mathbf{a})v_x^2 + 0.5f_{xy}(\mathbf{a})v_xv_y + 0.5f_{yy}(\mathbf{a})v_{y}^2$$

$$ = P_1(\mathbf{x}\,|\,\mathbf{a}) + v_x^2 + v_y^2$$

More specifically,  

$$ P_2(\mathbf{x}):=P_2\big(\mathbf{x}\,|\,(1,1)\big) = 2(x+y-1) + (x-1)^2 + (y-1)^2$$

```python
x, y = sp.symbols('x y')
P_2 = 2*(x+y-1) + (x-1)**2 + (y-1)**2; P_2.expand()
```
