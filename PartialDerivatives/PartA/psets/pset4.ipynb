The linear appproximation of $z$ at $P_0$ is:  

$$ P_1 = z_0 + \frac{x_0}{z_0}\,\Delta x + \frac{y_0}{z_0}\,\Delta y$$

**b)**  The ray is given by:  

$$ \mathbf{r} := \mathbf{r}(t) = t\langle x_0,y_0,z_0\rangle \quad\quad\text{where } t \ge 0$$

Or,  

$$ x = tx_0 \quad\quad y = ty_0 \quad\quad z=tz_0 $$

Via the above parametric equations, the equation that gives our cone, and substitution,  

$$ tz_0 = \sqrt{t^2x_0^2 + t^2y_0^2} = t\sqrt{x_0^2 + y_0^2} \quad\implies\quad z_0 = \sqrt{x_0^2 + y_0^2}$$

Thus, our ray fits on the cone.

Now, let $P_1:=z$.  

$$ tz_0 - z_0 = \frac{x_0}{z_0}\,(tx_0-x_0) + \frac{y_0}{z_0}\,(ty_0-y_0)$$

$$ z_0^2(t-1) = x_0^2(t-1) + y_0^2(t-1)$$

Dividing throughout by $t-1$ gives and equation that is consistent with the equation that gives our cone.

#### $\S\, 2\mathrm{B} : 3$  

In general, the hypotenuse $w$ of a right triangle with legs of length $x$ and $y$ is given by:  

$$ w = \sqrt{x^2 + y^2}$$

$$ w_x = \frac{x}{\sqrt{x^2+y^2}} \quad\quad\text{and}\quad\quad w_y = \frac{y}{\sqrt{x^2+y^2}}$$

The linear approximation of $w$ at $\mathbf{a}=(3,4)$ is given by:  

$$ P_1 = 5 + w_x(3,4)(x-3) + w_y(3,4)(y-4) = 5 + \frac{3}{5}\,(x-3) + \frac{4}{5}\,(y-4)$$

$$ w \approxeq P_1 \quad\text{where } \mathbf{x} = (x,y)\approxeq \mathbf{a}$$

Subtracting throughout the above equation by $w(3,4)=w_0$ gives:  

$$ \Delta w \approxeq \frac{3\Delta x + 4\Delta y}{5}$$

Assuming $\Delta x = \Delta y = 0.01$,  

$$ \Delta w \approxeq \frac{7}{5}\,0.01 = \frac{7}{500}$$

#### $\S\,2\mathrm{B} : 6$  

In general, the volume $V$ of a cylinder of radius $r$ and height $h$ is given by:  

$$ V = V(r,h) = V(\mathbf{x}) = \pi r^2h \quad\quad\text{where }\mathbf{x} = (x_1,x_2) = (r,h)$$

The linear approximation of $V$ at $\mathbf{a}=(a_r,a_h)=(2,3)$ is given by:  

$$ P_1(r,h) =P_1(\mathbf{x}) = V(\mathbf{a}) + V_r(\mathbf{a})(r-2) + V_h(\mathbf{a})(h-3)$$

The partial derivatives of $V$ are:  

$$ V_r = 2\pi r h \quad\quad\text{and}\quad\quad V_h = \pi r^2$$

Thus,  

$$ P_1(r,h) = 12\pi + 12\pi(r-2) + 4\pi(h-3)$$

$$ P_1(r,h) = 12\pi r + 4\pi h-24\pi = 4\pi(3r + h - 6)$$

Now, the task is to find a reasonable square centered at $\mathbf{a}$ wherein $\Delta P_1 = P_1(r,h)-P_1(2,3) \le 0.1$.

$$ \Delta P_1 = P_1(r,h)-P_1(2,3) = 12\pi \Delta r + 4\pi \Delta h \le 0.1$$

where,  

$$ \Delta r = r-2 \quad\quad\text{and}\quad\quad \Delta h = h-3$$

Whereas it is a square we are looking for therefore  

$$ \Delta r = \Delta h$$

Thus,  

$$ 12\pi\Delta r + 4\pi \Delta r = 16\pi \Delta r \le 0.1$$

$$ \Delta r = \Delta h \le \frac{1}{160\pi}$$
