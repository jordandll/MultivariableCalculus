#### Problem 1  

$$ R := R(w,r) = k\frac{w}{r^4}$$

$$ R_w = \frac{k}{r^4} \quad\quad\text{and}\quad\quad R_r = -4k\frac{w}{r^5}$$

**a)**  $$ \mathrm{d}R = R_w\mathrm{d}w + R_r\mathrm{d}r$$

**b)**  

$$ R_w = \frac{R}{w} \quad\quad\text{and}\quad\quad R_r = -R\frac{4}{r}$$

Thus,  

$$ \mathrm{d}R = \frac{R}{w}\mathrm{d}w  - R\frac{4}{r}\mathrm{d}r$$

$$ \frac{\mathrm{d}R}{R} = \frac{\mathrm{d}w}{w} - \frac{4\mathrm{d}r}{r}$$

**c)**  $r$ contributes more since the coefficient to it's relative change in the above equation is larger in magnitude than the coefficient to the relative change in the other variable.  In order to produce the greatest relative change in $R$, the changes to $w$ and $r$ should have opposite signs since the signs are their coefficents oppose each other.
