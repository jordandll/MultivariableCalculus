# Langrange Multipliers

## Problems 

#### Problem 1

Find the maximum and minimum values of $f(x, y) = x^2 + x + 2y^2$ on the unit circle.

$$ g(x,y) = x^2+y^2 = 1$$

$$ \nabla f = \langle 2x+1, 4y\rangle \quad\quad \nabla g = 2\langle x,y\rangle$$

The critical points of $f$ on the unit circle are given by:  

$$ \nabla f = \lambda \nabla g \quad\quad g(x,y)=1 \quad\implies$$

$$ \lambda 2x = 2x + 1 \quad\quad \lambda 2y = 4y $$

$$ \lambda = 2 \quad\implies\quad 4x = 2x+1 \quad\implies$$

$$ x  = 0.5 \quad\implies\quad y = \pm \sqrt{1-0.25} = \pm \frac{\sqrt{3}}{2}$$

At the critical points found above:  

$$ f\big(0.5,\pm 0.5\sqrt{3}\big) = 0.75 + 0.5\cdot 3 = \frac{3}{4} + \frac{3}{2} = \frac{9}{4}$$

At the boundaries,  

$$ f(0,\pm1) = 2 \quad\quad f(1, 0) = 2 \quad\quad f(-1,0) = 0$$

Thus, the min of $f$ is $0$ and the max is $2.25$.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = np.linspace(-1.125, 1.125,num=50)
X, Y = np.meshgrid(x,x)
f = X**2 + X + 2*Y**2

x = np.linspace(-1,1,num=200)
y_p = np.sqrt(1-x**2)
y_n = -y_p

fig, ax = plt.subplots()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

cs = ax.contour(X,Y,f,levels=12)
cb = fig.colorbar(cs)
ax.plot(x,y_p,x,y_n, color='k', ls='--')
```

#### Problem 2

Find the minimum and maximum values of $f(x, y) = x^2 − xy + y^2$ on the quarter circle
$x^2 + y^2 = 1$, $x, y \ge 0$.

$$ \nabla f = \langle 2x-y, 2y-x\rangle \quad\quad \nabla g = 2\langle x,y\rangle$$

The critical points of $f$ on the quarter circle are given by:  

$$\nabla f= \lambda \nabla g \quad\quad g(x,y)=1 $$

$$ \lambda 2x = 2x-y \quad\quad \lambda 2y = 2y -x$$

$$ 2\lambda = 2 - \frac{y}{x} = 2 - \frac{x}{y} \quad\implies\quad \frac{y}{x} = \frac{x}{y}$$

$$\implies\quad  x^2 = y^2 \quad\implies\quad x^2 = 0.5 \quad\implies$$

$$ x = \pm \frac{1}{\sqrt{2}} \quad\implies\quad x = y = \frac{1}{\sqrt{2}}$$

At the critical point found above:  

$$ f\big(\sqrt{0.5},\sqrt{0.5}\big) = 0.5$$

And, at the boundaries:  

$$ f(1,0) = 1 \quad\quad f(0,1) = 1$$

Thus,  

$$ f_\min = 0.5\quad\quad f_\max = 1$$

```python
x = np.linspace(-0.06, 1.125, num=50)
X, Y = np.meshgrid(x,x)
f = X**2 - X*Y + Y**2

x = np.linspace(0,1,num=200)
y = np.sqrt(1-x**2)

fig, ax = plt.subplots()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

cs = ax.contour(X,Y,f,levels=12)
cb = fig.colorbar(cs)
ax.plot(x,y,color='k',ls='--')
```

```python
1/np.sqrt(2)
```
