```python
x = np.linspace(-5,5,num=50)
y_max = np.sqrt(5**2-3)
y = np.linspace(-y_max,y_max,num=50)
X, Y = np.meshgrid(x,y)

f = X**2-Y**2
g = X**2+Y**2

fig, ax = plt.subplots()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.grid()

csf = ax.contour(X,Y,f,levels=[-3,-2,-1,1,2,3])
csg = ax.contour(X,Y,g,linestyles='dashed',levels=[1,2,3])
cb = fig.colorbar(csf)
```

$$ f(x,y) = x^2-y^2 \quad\quad\text{and}\quad\quad g(x,y) = x^2+y^2$$

$$ \nabla f = 2\langle x,-y\rangle \quad\quad\text{and}\quad\quad \nabla g = 2\langle x,y\rangle$$

**a)**  $$ \lambda = \{1,-1\}$$

The gradients are proportional (parallel?) if and only if $\lambda=1$ and $P=(x,0)$ or $\lambda=-1$ and $P=(0,y)$.

\begin{equation}\tag{1.1}
    \nabla f = \lambda \nabla g  \quad\iff\quad \nabla f \propto \nabla g \quad\iff\quad \lambda = 1\;\land\;P=(x,0)\;\lor\;\lambda = -1\;\land\;P=(0,y)
\end{equation}

Note that when the gradients are proportional they are parallel, except when $P=(0,0)$.

Just by looking at the above figure that contains the contour plots of both functions, it is apparent that, under the aforementioned constraint, the min of $f$ is equal to $-3$ and occurs at $\big(0,\pm\sqrt{3}\big)$, while the max of $f$ is equal to $3$ and occurs at $\big(\pm\sqrt{3},0\big)$.  This is where the contour lines given by $f(x,y)$ are tangential to the circle given by the constraint.

Now, without looking at the figure, the critical points of $f$ under the constraint are given by:  

$$ \langle x,-y\rangle = \lambda \langle x,y\rangle \quad\quad x^2+y^2 = 3$$

The extremum of $f$ exist at these critical points, since there are no boundary cases.  The solution(s) to the above system, of course, rest on the constraint curve and are where the gradients of $f$ and $g$ are parallel and the level curve(s) of $f$ and the constraint curve are tangential.

**c)**  What do the two values of $\lambda$ correspond to in terms of the pairs of solution points?
Do the gradients of $f$ and $g$ point in the same or the opposite direction at the contact

There are two pairs of critical points:  One for $\lambda=1$, denoted as '$\mathbf{a}$'; And one for $\lambda = -1$, denoted as '$\mathbf{b}$'.  

$$ \nabla f(\mathbf{a}) = \nabla g(\mathbf{a}) \quad\quad\text{where } g(\mathbf{a}) = 3 \quad\implies$$

$$ \langle a_x, -a_y\rangle = \langle a_x,a_y\rangle \quad\quad\text{where } |\mathbf{a}|^2 = 3$$

Via (1.1) and the fact that $\nabla f(\mathbf{a}) \parallel \nabla g(\mathbf{a})$,  

$$ \mathbf{a} = \langle a_x,0\rangle$$

And, via the constraint and the above equation,  

$$ a_x^2 = 3 \quad\therefore\quad a_x = \pm\sqrt{3}$$

Thus,  

$$ \mathbf{a} = \big\langle\pm\!\sqrt{3},0\big\rangle$$

The fact $\nabla f(\mathbf{a})=\nabla g(\mathbf{a})$ implies that the two gradients point in the same direction.

Furthermore,  

$$  \nabla f(\mathbf{b}) = -\nabla g(\mathbf{b}) \quad\quad\text{where } g(\mathbf{b}) = 3 \quad\implies$$

$$\langle b_x, -b_y\rangle = -\langle b_x,b_y\rangle \quad\quad\text{where } |\mathbf{b}|^2 = 3$$

Via (1.1) and the fact that $\nabla f(\mathbf{b}) \parallel \nabla g(\mathbf{b})$,  

$$ \mathbf{b} = \langle 0,b_y\rangle$$

And, via the constraint and the above equation,  

$$ b_y^2 = 3 \quad\therefore\quad b_y = \pm\sqrt{3}$$

Thus,  

$$ \mathbf{b} = \big\langle0,\pm\sqrt{3}\big\rangle$$

The fact $\nabla f(\mathbf{b})=-\nabla g(\mathbf{b})$ implies that the two gradients point in the opposite direction when $\lambda = -1$.

**a)** Determine what choice of $I_1$ and $I_2$ will minimize energy loss and hence determine
what the currents will be along the two paths. (Alternatively, if you already are
familiar with resistors in parallel and current flows, verify that the currents $I_1$ and $I_2$
do in fact minimize energy loss).

Let $f(I_1,I_2)$ and $g(I_1,I_2)$ be defined as:  

$$ P = f(I_1,I_2) := R_1I_1^2 + R_2I_2^2 \quad\quad\text{and}\quad\quad I = g(I_1,I_2) := I_1 + I_2$$

The critical point of $f$ on the level curve given by the constrainment involving $g$ are given by:  

$$ \nabla f = \lambda \nabla g \quad\quad\text{and}\quad\quad I_1+I_2=I$$

$$ \nabla f = 2\langle R_1I_1, R_2I_2\rangle\quad\quad\text{and}\quad\quad \nabla g = \langle 1,1\rangle$$

Thus,  

\begin{equation}\tag{2.1}
    R_1I_1 = R_2I_2
\end{equation}

Via the above equation and the constrainment equation and substitution,  

$$ I_1 + \frac{R_1}{R_2}\,I_1 = I \quad\therefore\quad I_1 = \frac{R_2+R_1}{R_2} = I \quad\therefore$$

\begin{equation}\tag{2.2a}
    I_1 = \frac{R_2}{R_2+R_1}\,I
\end{equation}

And, via the above equation and equation (2.1) and substitution,  

\begin{equation}\tag{2.2b}
    I_2 = \frac{R_1}{R_2+R_1}\,I
\end{equation}

And, since the max value of $P$ occurs at the boundaries, the critical point given by equations (2.2a,b) are where the min value of $P$ exist.

For those that know, equation (2.2) is also given by KVL and Ohm's Law, while the constrainment equation is given by KCL.

**b)** Suppose instead we had three resistors in parallel. In terms of $R_1$, $R_2$, and $R_3$
determine the values of $I_1$, $I_2$, and $I_3$ which minimize energy loss.

Let $f(I_1,I_2,I_3)$ and $g(I_1,I_2,I_3)$ be defined as:  

$$ P = f(I_1,I_2,I_3) := R_1I_1^2 + R_2I_2^2 + R_3I_3^2\quad\quad\text{and}\quad\quad I = g(I_1,I_2,I_3) := I_1 + I_2 + I_3$$

The critical point of $f$ on the level surface given by the constrainment involving $g$ are given by:  

$$ \nabla f = \lambda \nabla g \quad\quad\text{and}\quad\quad I_1+I_2 + I_3=I$$

$$ \nabla f = 2\langle R_1I_1, R_2I_2, R_3I_3\rangle\quad\quad\text{and}\quad\quad \nabla g = \langle 1,1,1\rangle$$

Thus,  

\begin{equation}\tag{2.3}
    R_1I_1 = R_2I_2 = R_3I_3
\end{equation}

Via the above equation and the constrainment equation and substitution,  

$$ I_1 + \frac{R_1}{R_2}\,I_1 + \frac{R_1}{R_3}\,I_1 = I \quad\therefore\quad I_1\,\frac{R_2R_3+R_1R_3 + R_2R_1}{R_2R_3} = I \quad\therefore$$

\begin{equation}\tag{2.4a}
    I_1 = \frac{R_2R_3}{R_2R_3+R_1R_3+R_2R_1}\,I
\end{equation}

Via the above equation and equation (2.3) and substitution,  

\begin{equation}\tag{2.4b}
    I_2 = \frac{R_1R_3}{R_2R_3+R_1R_3+R_2R_1}\,I
\end{equation}

and  

\begin{equation}\tag{2.4c}
    I_3 = \frac{R_1R_2}{R_2R_3+R_1R_3+R_2R_1}\,I
\end{equation}

Finally, since the max value of $P$ occurs at the boundaries, the min value of $P$ occurs at the critical point identified by equations (2.4a,b,c).
