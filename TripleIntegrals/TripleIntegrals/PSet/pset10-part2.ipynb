### Problem 1  

Let $\mathcal{G}$ be the solid region in $3$-space which lies inside the surface $x^2 + (y âˆ’ 1)^2 = 1$, 
above $z = 0$, and below the surface $z = \sqrt{x^2 + y^2}$ .  

a) Find the volume of $\mathcal{G}$.  

b) Find the $z$-coordinate of the centroid of $\mathcal{G}$

**a)**  The surface given by:  

$$ x^2 + (y-1)^2 = 1 \quad\quad\text{or}\quad\quad x^2 + y^2 -2y = 0$$

is a hollow cylinder with a radius of $1$ and a center line through $(0,1)$ and a height of infinity.  It is also given by, in cylindrical coordinates,  

$$ r^2  = 2r\sin\theta \quad\quad\text{or}\quad\quad r := r(\theta) = 2\sin\theta$$

The volume is:  

$$ V = \iiint_{\mathcal{G}}\,\mathrm{d}V = \iint_R\int_0^r \mathrm{d}z\,\mathrm{d}A$$

where $R$ is bounded by $r(\theta)$.  Therefore,  

$$ V = \int_0^\pi\int_0^{2\sin\theta} r^2\,\mathrm{d}r\,\mathrm{d}\theta$$

$$ = \frac{8}{3}\int_0^\pi \sin^3\theta\,\mathrm{d}\theta$$

$$\int\sin^3\theta\,\mathrm{d}\theta =  g(\theta) = -\frac{2}{3}\cos^3\theta-\cos\theta\sin^2\theta + K$$

Finally,  

$$ V = \frac{8}{3}\big(g(\pi)-g(0)\big) = \frac{8}{3}\left(\frac{2}{3} + \frac{2}{3}\right) = \frac{32}{9}$$

```python
32/9
```

**b)**  $$ \overline{z} = \frac{1}{V}\iiint_{\mathcal{G}} z\,\mathrm{d}V = \frac{9}{32}\iint_R\int_0^r z\,\mathrm{d}z\,\mathrm{d}A$$

$$ = \frac{9}{64}\int_0^\pi\int_0^{2\sin\theta}r^3\,\mathrm{d}r\,\mathrm{d}\theta$$

$$ = \frac{9}{16}\int_0^\pi\sin^4\theta\,\mathrm{d}\theta$$

$$ \sin^4\theta = \big(1-\cos^2\theta\big)^2 = 1 - 2\cos^2\theta + \cos^4\theta$$

$$ = -\cos2\theta + \frac{1}{4}\big(\cos2\theta + 1\big)^2$$

$$ = \frac{1}{4}\big(\cos^2 2\theta + 2\cos2\theta + 1\big) - \cos2\theta$$

$$ = \frac{\cos^2 2\theta + 1}{4} - \frac{\cos 2\theta}{2} = \frac{\cos 4\theta + 1}{8} + \frac{1}{4} - \frac{\cos2\theta}{2}$$

$$ = \frac{\cos 4\theta + 3 - 4\cos 2\theta}{8}$$

Finally,  

$$ \overline{z} = \frac{27\pi}{128}$$

### Problem 3  

The work done is:  

$$ W = \iiint_D f(x,y,z)\,\mathrm{d}V$$

where $D$ is the region occupied by the body of water stored in the vessel and  

$$ f(x,y,z) = \delta g z \quad\quad \delta = 1\,\frac{\mathrm{kg}}{\mathrm{m}^3} \quad\quad g = 9.8\,\frac{\mathrm{m}}{\mathrm{s}^2}$$

$$ f(x,y,z) = 9.8\frac{\mathrm{kg}}{\mathrm{m}^2\mathrm{s}^2}\cdot z$$

Therefore,  

$$ W = 9.8\,\frac{\mathrm{kg}}{\mathrm{m}^2\mathrm{s}^2}\int_{x_1}^{x_2}\int_0^{200\mathrm{m}}\int_{0.01 x^4\mathrm{m}^{-3}}^{81\mathrm{m}}z\,\mathrm{d}z\,\mathrm{d}y\,\mathrm{d}x$$

where,  

$$ \frac{x^4}{100\mathrm{m}^3} = 81\mathrm{m} \quad\text{or}\quad x^4 = 8100\mathrm{m}^4 \quad\text{or}\quad\quad x = \pm 3\sqrt{10}\,\mathrm{m}$$

$$ x_1 = -3\sqrt{10}\,\mathrm{m} \quad\quad\text{and}\quad\quad x_2 = 3\sqrt{10}\,\mathrm{m}$$

Therefore,  

$$ W = \frac{9.8}{2}\,\frac{\mathrm{kg}}{\mathrm{m}^2\mathrm{s}^2}\int_{x_1}^{x_2}\int_0^{200\mathrm{m}}\left(81^2\mathrm{m}^2 - \frac{x^8}{10000\mathrm{m}^6}\right)\,\mathrm{d}y\,\mathrm{d}x$$

$$ = 81^2 \cdot 980\,\frac{\mathrm{m}\cdot\mathrm{kg}}{\mathrm{s}^2}\int_{x_1}^{x_2}\mathrm{d}x - \frac{9.8}{100}\,\frac{\mathrm{kg}}{\mathrm{m}^7\mathrm{s}^2}\int_{x_1}^{x_2}x^8\,\mathrm{d}x$$

$$ = 2\cdot 3^9\sqrt{10}\cdot 980\,\mathrm{J} - \frac{9.8}{450}\cdot 3^9\sqrt{10^9},\mathrm{J}$$

$$ = 3^9\sqrt{10}\cdot 9.8\left(200 - \frac{10^4}{450}\right)\,\mathrm{J}$$

$$ = 3^9 100\sqrt{10} \cdot 9.8\left(2-\frac{10}{45}\right)\,\mathrm{J}$$

$$ = 3^9 100\sqrt{10} \cdot 9.8 \left(2 - \frac{2}{9}\right)\,\mathrm{J}$$

$$= 3^7 2^4 100 \sqrt{10}\cdot 9.8 \,\mathrm{J}$$
