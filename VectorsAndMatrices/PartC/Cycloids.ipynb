#### Problem 1

1. A disk of radius $2 \mathrm{cm}$ slides at a speed of $12\sqrt{2} \mathrm{cm}/\mathrm{sec}$ in the direction of $\langle 1, 1\rangle$. As it slides it spins counterclockwise at $3$ revolutions per second. Measuring time in seconds, at time $t = 0$ the diskâ€™s center is at the origin $(0,0)$.  

Find parametric equations for the trajectory of the point $P$ on the edge of the disk, which
is initially at $(2, 0)$.

Let $a$ and $C$ denote the radius and center of the disk respectively.  Now let $\mathbf{P}:=\overrightarrow{OP}$ and $\mathbf{C}:=\overrightarrow{OC}$.  It now follows that $\mathbf{P}(t)$ and $\mathbf{C}(t)$ are the position vectors of $P$ and $C$ as a function of time respectively.  

$$ \mathbf{P} = \mathbf{C} + \overrightarrow{CP}$$

Note that the unit of measurement of the vector plane is the centimeter ($\mathrm{cm}$).  Thus, the $x$-compliment of $\mathbf{P}(0)$ is equal to $2\mathrm{cm}$, which is equal to the radius.  

$$ P(0) = (a, 0) \quad\quad C(0) = (0,0)$$

Now, let $\theta$ denote the degree to which the disk has spun since $t=0$.  It now follows that:  

$$ \theta = \theta(t) = 6\pi t\quad\quad \theta(0)=0 \quad\quad \frac{\mathrm{d}\theta}{\mathrm{d}t} = \theta'(t) = 6\pi$$

$$ \overrightarrow{CP} = a\langle \cos\theta, \sin\theta\rangle $$

Now, let $\mathbf{\hat{u}}:=\langle 1,1\rangle$.  The component of $\mathbf{C}$ in the direction of $\mathbf{\hat{u}}$ is given by:  

$$ \frac{\mathbf{C}\cdot\mathbf{\hat{u}}}{|\mathbf{\hat{u}}|} = \mathbf{C}\cdot\mathbf{u}\quad\quad\text{where }\mathbf{u} = \frac{\langle 1,1\rangle}{\sqrt{2}}$$

And,  

$$ \mathbf{C} = \mathbf{C}(t) = \big(12\sqrt{2} t-a\theta\big)\mathbf{u} \;\therefore\; |\mathbf{C}| = |\mathbf{C}|(t) =  \big|12\sqrt{2} t - a\theta\big|$$

Ultimately, $\mathbf{C}$ is given by:  

$$ \mathbf{C} = \mathbf{C}(t) = \frac{12\sqrt{2} t - 12\pi t}{\sqrt{2}}\,\langle 1, 1\rangle = \big(12 t - \sqrt{2}6\pi t\big)\langle 1,1\rangle$$

$$  = 6t\big(2 - \sqrt{2}\pi\big)\langle 1, 1\rangle$$

Finally,  

$$ \mathbf{P} = \mathbf{P}(t) = 6t\big(2-\sqrt{2}\pi\big)\langle 1,1\rangle + 2\langle\cos6\pi t, \sin6\pi t\rangle$$

$$ P_x= x = 6t\big(2-\sqrt{2}\pi\big) + 2\cos6\pi t $$

$$ P_y  = y = 6t\big(2-\sqrt{2}\pi\big) + 2\sin6\pi t $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,1,num=200)
theta = 6*np.pi*t
y = 6*t*(2-np.sqrt(2)*np.pi) + 2*np.sin(theta)
x = 6*t*(2-np.sqrt(2)*np.pi) + 2*np.cos(theta)

# x = 12*t + 2*np.cos(theta)
# y = 12*t + 2*np.sin(theta)

fig, ax = plt.subplots()
ax.set_title('Cycloid', size=14)
ax.set_xlabel(r'x-axis ($\mathrm{cm}$)')
ax.set_ylabel(r'y-axis ($\mathrm{cm}$)')

ax.plot(x,y)
```
